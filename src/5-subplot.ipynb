### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the plots
x = np.arange(1, 6)
y = x * 2

categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]

scatter_x = [1, 2, 3, 4, 5]
scatter_y = [2, 4, 6, 8, 10]

pie_labels = ['Python', 'Java', 'C++', 'JavaScript']
pie_sizes = [40, 25, 20, 15]

# Create the 2x2 subplot layout
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Line Plot
axes[0, 0].plot(x, y, marker='o', linestyle='-', color='b')
axes[0, 0].set_title("Line Plot")
axes[0, 0].set_xlabel("X values")
axes[0, 0].set_ylabel("Y values")
axes[0, 0].grid(True)

# Bar Chart
axes[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'orange'])
axes[0, 1].set_title("Bar Chart")
axes[0, 1].set_xlabel("Categories")
axes[0, 1].set_ylabel("Values")
axes[0, 1].grid(axis='y', linestyle='--', alpha=0.7)

# Scatter Plot
axes[1, 0].scatter(scatter_x, scatter_y, color='purple', marker='o')
axes[1, 0].set_title("Scatter Plot")
axes[1, 0].set_xlabel("X values")
axes[1, 0].set_ylabel("Y values")
axes[1, 0].grid(True)

# Pie Chart
axes[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', startangle=140, colors=['blue', 'red', 'green', 'orange'])
axes[1, 1].set_title("Pie Chart")

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```
